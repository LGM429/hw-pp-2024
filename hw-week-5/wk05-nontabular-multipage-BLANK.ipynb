```python
## below code should scrape the info I need. Then I'll have to zip together and make a dateframe.
```

```python
artists_fl_m = []
albums_fl_m = []
sales_fl_m = []
links_fl_m = []

for url_number in range(1,total_pages):
    print(f"Scraping link {counter} of {total_pages}")
    counter+=1 ## increment counter
    link = f"{base_url}{url_number}"
    response = r.get(link)
    
    try:
        soup_m = BeautifulSoup(response.text, "html.parser")
        artists_m = soup_m.find_all("div", class_ = "artist")
        albums_m = soup_m.find_all("div", class_ = "album")
        sales_m = soup_m.find_all("div", class_ = "sales")

        for artist in artists_m:
            artists_fl_m.append(artist.get_text())

        for album in albums_m:
            albums_fl_m.append(album.get_text())

        for number in sales_m:
            sales_fl_m.append((number.get_text()).replace("Sales: ",""))

        for webpage in albums_m:
            links_fl_m.append(webpage.find("a").get("href"))

    except:
        print(f"Something is wrong with {link}!")
        error_links.append(link)
        
    finally:
        ## let's not forget to snooze
        snooze = randrange(5,7)
        print(f"snoozing for {snooze} seconds before scraping next link.")
        time.sleep(snooze)
        

    print("Done scraping")
```
