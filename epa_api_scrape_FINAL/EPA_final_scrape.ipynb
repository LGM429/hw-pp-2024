```python

base_url = "https://aqs.epa.gov/data/api/quarterlyData/byCounty?email=leo.miranda84@journalism.cuny.edu&key=ecrukit55&param=43824&bdate="
dates = ["20180101&edate=20181231", "20190101&edate=20191231", "20200101&edate=20201231", "20210101&edate=20211231", "20220101&edate=20221231", "20230101&edate=20231231"]
end_url = "&state=36&county=047"
```

```python
endpoints_fl = []
for date in dates:
    endpoint = base_url+date+end_url
    endpoints_fl.append(endpoint)
    
```

```python
## That worked with just one record path. I dont think I need the metadata
```

```python
df_list = []
for endpoint in endpoints_fl:
    response = requests.get(endpoint)
    aqs_data = response.json()
    df = pd.json_normalize(aqs_data, record_path = ["Data"])
    df_list.append(df)
```

```python
len(df_list)
```

```python
df_all = pd.concat(df_list, ignore_index = True)
df_all
```
