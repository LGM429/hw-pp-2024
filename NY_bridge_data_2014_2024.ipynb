```python
def convert_to_date(code):
    code = str(code)
    if len(code) == 3:
        month = code[:2].zfill(2)  # Ensure month is 2 digits
        year = '20' + code[2:]     # Assume year is in the 2000s
    elif len(code) == 4:
        month = code[:2]
        year = '20' + code[2:]
    else:
        raise ValueError(f"Invalid code length: {code}")

    # Return the date formatted as 'YYYY-MM-01' for correct datetime conversion
    return f"{year}-{month}-01"

# Apply the conversion and convert to datetime
df_2['DATE_OF_INSPECT_090'] = pd.to_datetime(df_2['DATE_OF_INSPECT_090'].apply(convert_to_date))

# Now format as '%m-%y' (month-year format)
df_2['DATE_OF_INSPECT_090'] = df_2['DATE_OF_INSPECT_090'].dt.strftime('%m-%y')
```
