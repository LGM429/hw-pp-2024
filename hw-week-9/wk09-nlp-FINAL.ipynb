````python
def tokenize(file_path): ## AI helped with this
    with open(file_path, "r") as my_text:
        text = my_text.read()
    return text
````

````python
tokenize("senate-hearing.txt")
````

````python
s_doc = nlp(tokenize("senate-hearing.txt"))
````

````python
s_doc
````

````python
type(s_doc)
````

````python
entities = [word.text for word in s_doc.ents] #gives entity/word
labels = [word.label_ for word in s_doc.ents] #gives label
meanings = [spacy.explain(word.label_) for word in s_doc.ents] #gives meaning
````
